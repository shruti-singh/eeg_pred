```python
with open('data/sentiment_labels_task1.csv', 'r') as f:
    lines = f.readlines()
```

```python
sent_sents = []

for line in lines:
    proc_line = line.replace(";CONTROL;", ";;")
    proc_line = proc_line.replace(";;", ";")
    sent_sents.append(proc_line)
```

```python
sentiment_dict = {}
rev_sentiment_dict = {}

for s in sent_sents[2:]:
    s = s.strip()
    ps = s.split(";")
    if len(ps) != 3:
        print(s)
        s = s.replace(";", "")
        ps = s.split('"')
        sentiment_dict[ps[0]] = {"sent": ps[1], "label": int(ps[2])}
        rev_sentiment_dict[ps[1]] = {"id": ps[0], "label": int(ps[2])}
        continue
    sentiment_dict[ps[0]] = {"sent": ps[1], "label": int(ps[2])}
    rev_sentiment_dict[ps[1]] = {"id": ps[0], "label": int(ps[2])}
```

```python
len(rev_sentiment_dict) == len(sentiment_dict)
```

```python
import pickle
```

```python
with open('data/sentiment_gold.pkl', 'wb') as f:
    pickle.dump(sentiment_dict, f)
    
with open('data/rev_sentiment_gold.pkl', 'wb') as f:
    pickle.dump(rev_sentiment_dict, f)
```
